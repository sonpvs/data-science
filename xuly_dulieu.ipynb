```python
# Xử lý dữ liệu lỗi
df.fillna(df.mean())
```

```python
# Phân nhóm Tuoi
# Tuổi : 22 đến  29 =  1;   30 đến  45 = 2;   Trên 45 = 3

df['NhomTuoi']=0

def nhtuoi(a):
    if (a >= 22) & (a<=29):
       return 1
    elif (a >= 30) & (a<=45):
       return 2
    else:
       return 3


# Chạy từng dòng để nhóm độ tuổi
for row in df.index:
    c= df.iloc[row]['TUỔI']
    df.iloc[row, df.columns.get_loc('NhomTuoi')] = nhtuoi(c)
```

```python
# Phân nhóm Hoc van
# Học vấn: 5 đến  7 = 0;   9 đến 15 = 1

df['NhomHocvan']=0

def nh_Hocvan(a):
    if (a >= 5) & (a<=8):
       return 0
    elif (a >= 9) & (a<=15):
       return 1

# Chạy từng dòng để nhóm học vấn
for row in df.index:
    c= df.iloc[row]['HỌC VẤN']
    df.iloc[row, df.columns.get_loc('NhomHocvan')] = nh_Hocvan(c)
```

```python
# Phân nhóm Tuổi kết hôn
#Tuổi kết hôn: 16 đến 18 = 0; 19 đến  27 = 1

df['NhomTuoiKH']=0

def nh_TuoiKH(a):
    if (a >= 16) & (a<=18):
       return 0
    elif (a >= 19) & (a<=27):
       return 1

for row in df.index:
    c= df.iloc[row]['TUỔI KẾT HÔN']
    df.iloc[row, df.columns.get_loc('NhomTuoiKH')] = nh_TuoiKH(c)
```

```python
# Phân nhóm Số con
# Số con: 1 đến 2 = 1; 3 đến 5 = 0 

df['NhomSocon']=0

def nh_TuoiSocon(a):
    if (a >= 1) & (a<=2):
       return 1
    elif (a >= 3) & (a<=5):
       return 0

for row in df.index:
    c= df.iloc[row]['SỐ CON']
    df.iloc[row, df.columns.get_loc('NhomSocon')] = nh_TuoiSocon(c)
```
