```python
list_cot1 =["NhomTuoi", "NhomHocvan", "NhomTuoiKH", "HÔN NHÂN", "NhomSocon", "NGHỀ NGHIỆP", "THU NHẬP", "TB QUYỀN QUYẾT ĐỊNH SẢN XUẤT", "TB QUYẾT ĐỊNH TÍN DỤNG", "TB QUYẾT ĐỊNH CHI TIÊU GIA ĐÌNH", "TB QUYỀN SỞ HỮU TÀI SẢN", "TB QUYỀN THAM GIA XÃ HỘI", "TB QUYỀN SD THỜI GIAN", "TB BẤT BĐ GIỚI", "NHỊ PHÂN BĐG"]
```

```python
df01 = pd.DataFrame(df,columns=list_cot1)
```

```python
df01.head()
```

```python
df01.rename(columns={"NhomTuoi": "NHÓM TUỔI", "NhomHocvan": "NHÓM HỌC VẤN","NhomTuoiKH": "NHÓM TUỔI KẾT HÔN", "NhomSocon": "NHÓM SỐ CON"}, inplace=True)
```

```python
df01.head()
```

```python
df01.fillna(df01.mean)
```

```python
from sklearn.metrics import mean_squared_error, mean_squared_log_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
train, test = train_test_split(df01, test_size=0.3, shuffle=True)
training_features = ['NHÓM TUỔI','NHÓM HỌC VẤN','NHÓM TUỔI KẾT HÔN','HÔN NHÂN','NHÓM SỐ CON','NGHỀ NGHIỆP', 'THU NHẬP']
target = 'NHỊ PHÂN BĐG'
train_X = train[training_features]
train_Y = train[target]
test_X = test[training_features]
test_Y = test[target]
```

```python
logreg = LogisticRegression()
```

```python
#df01.isna
# using dropna() function    
df01.dropna(how = 'all')
```

```python
train_Y = np.nan_to_num(train_Y)
```

```python
rfe = RFE(logreg, 15)             
rfe = rfe.fit(train_X, train_Y)
```

```python
print(rfe)
```

```python
test_X = np.nan_to_num(test_X)
test_Y = np.nan_to_num(test_Y)

score = rfe.score(test_X, test_Y)
print('Accuracy Score: ', score)
```

```python
# Predict class labels on a test data
LR1_pred_labels = rfe.predict(test_X)
```

```python
from sklearn.metrics import classification_report # for model evaluation metrics
print('--------------------------------------------------------')
# Look at classification report to evaluate the model
print(classification_report(test_Y, LR1_pred_labels))
```
